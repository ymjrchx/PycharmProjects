```python
seqOp=(lambda x,y:(x[0]+y,x[1]+1))
combOp=(lambda x,y:(x[0]+y[0],x[1]+y[1]))
x=sc.parallelize([1,2,3,4,5,6],2)
print(x.glom().collect())#返回RDD每个分区中的所有元素并合并到一个列表中
```

```python
y=x.aggregate((1,2),seqOp,combOp)#使用给定的组合函数和,聚合每个分区的元素，然后聚合所有分区的结果
print(y)
#计算过程如下：
#（1，2）--》（1+1，2+1）-->(2+2,3+1)-->(4+3，4+1)-->(7,5)；
#（1，2）--》（1+4，2+1）-->(5+5,3+1)-->(10+6，4+1)-->(16,5)；
#（1，2）--》（1+7，2+5）-->(8+16,7+5)-->(24，12)；
```

```python
y=x.aggregate((0,0),seqOp,combOp)
print(y)
#（0，0）--》（0+1，0+1）-->(2+1,1+1)-->(3+3，2+1)-->(6,3)；
#（0，0）--》（4+0，0+1）-->(4+5,1+1)-->(9+6，2+1)-->(15,3)；
#（0，0）--》（0+6，0+3）-->(6+15,3+3)-->(21，6)；
```
